```python
import pandas as pd
df = pd.read_csv("rounds.csv")
df.loc[:, "Survival_Time"] /= 60
df = df.rename({"Survival_Time": "Survival_Time (min)"}, axis=1)
```

```python
df.sample(5)
```

```python
import numpy as np
pd.DataFrame(np.unique(df.Model.to_numpy(), return_counts=True)).T
```

```python
df.sort_values('Survival_Time (min)', ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
d = df.sort_values('Survival_Time (min)', ascending=False).iloc[1:]
ax = sns.boxplot(d, y='Model', x='Survival_Time (min)')
sns.despine(offset=25, trim=True)
```
